### the following processed data comes from GSE49618:

```python
import methylcheck
import pandas as pd
from pathlib import Path
path = Path('~/GEO/test_pipeline/GSE49618').expanduser()
```

```python
beta_df = pd.read_pickle(Path(path,'beta_values.pkl'))
# OR
beta_df = methylcheck.load(Path(path,'beta_values.pkl'))
beta_df.head()
```

```python
m_df = pd.read_pickle(Path(path,'m_values.pkl'))
m_df.head()
```

```python
meta_df = pd.read_pickle(Path(path,'sample_sheet_meta_data.pkl'))
meta_df.head()
```

## these additional (optional) types of data are used for quality control only
(whereas m_values and beta_values are used for experiment analysis)

```python
# the structure is different for control -- a dictionary of dataframes, with sample names as keys and dataframes as values
controls = pd.read_pickle(Path(path,'control_probes.pkl'))
list(controls.values())[0].head()
```

```python
nmeth_df = pd.read_pickle(Path(path,'noob_meth_values.pkl'))
nmeth_df.head()
```

```python
nunmeth_df = pd.read_pickle(Path(path,'noob_unmeth_values.pkl'))
nunmeth_df.head()
```

```python
meth_df = pd.read_pickle(Path(path,'meth_values.pkl'))
meth_df.head()
```

```python
unmeth_df = pd.read_pickle(Path(path,'unmeth_values.pkl'))
unmeth_df.head()
```

## methylcheck plots

```python
methylcheck.sample_plot(beta_df)
```

## Control Probe Plots
The following functions are wrapped by the single function `methylcheck.run_qc(<path-to-files>)`.

```python
methylcheck.qc_signal_intensity(meth=meth_df, unmeth=unmeth_df)
methylcheck.qc_signal_intensity(meth=nmeth_df, unmeth=nunmeth_df)
# NOOB_meth and NOOB_unmeth have slightly different values, but should give similar results here.
```

```python
methylcheck.plot_M_vs_U(meth=meth_df, unmeth=unmeth_df)
```

```python
methylcheck.plot_beta_by_type(beta_df, 'all')
```

```python
methylcheck.plot_controls(controls, 'all')
```

```python
methylcheck.get_sex((meth_df, unmeth_df), plot=True) # pass in (meth_df, unmeth_df) as a pair of dataframes.
```

## if you want to run all QC functions and plots shown above, there is a convenience function, `run_qc()`
Just specify the path to the methylprep processed files.

```python
methylcheck.run_qc('.')
```
